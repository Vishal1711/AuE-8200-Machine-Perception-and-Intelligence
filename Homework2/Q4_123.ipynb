```python
import open3d as o3d
import numpy as np

seg_name = 'C:\\Users\\jadha\\Machine Perception and Intelligence\\NuScenes\\data\\sets\\nuscenes\\samples\\LIDAR_TOP\\n008-2018-08-01-15-16-36-0400__LIDAR_TOP__1533151617397050.pcd.bin'
seg=np.fromfile(seg_name, dtype=np.uint8)

color = np.zeros([len(seg), 3])
color[:, 0] = seg/32
color[:, 1] = seg/32
color[:, 2] = seg/32


pcd_name='C:\\Users\\jadha\\Machine Perception and Intelligence\\NuScenes\\data\\sets\\nuscenes\\samples\\LIDAR_TOP\\n008-2018-08-01-15-16-36-0400__LIDAR_TOP__1533151617397050.pcd.bin'
scan=np.fromfile(pcd_name, dtype=np.float32)
print(len(scan))

points = scan.reshape((-1, 5))[:, :4]
print(points)

print('\nHeight\n')
intensity = points[:,2]
print(intensity)

print('\nIntesnity\n')
intensity = points[:,3]
print(intensity)

pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points[:, :3])
pcd.colors = o3d.utility.Vector3dVector(color)

o3d.visualization.draw_geometries([pcd])
```

```python
import open3d as o3d
import numpy as np
import os.path as osp

from nuscenes.nuscenes import NuScenes
from nuscenes.utils.data_classes import RadarPointCloud

# Load the nuScenes dataset (mini-split, in this case).
nusc = NuScenes(dataroot='C:\\Users\\jadha\\Machine Perception and Intelligence\\NuScenes\\data\\sets\\nuscenes')

sample_record = nusc.sample[100]
pointsensor_token = sample_record['data']['RADAR_FRONT']
pointsensor = nusc.get('sample_data', pointsensor_token)
pcl_path = osp.join(nusc.dataroot, pointsensor['filename'])

pc = RadarPointCloud.from_file(pcl_path)
pc = pc.points.T

print('x y z dyn_prop id rcs vx vy vx_comp vy_comp is_quality_valid ambig_state x_rms y_rms invalid_state pdh0 vx_rms vy_rms')
print(pc)
x = pc[:,0]
y = pc[:,1]
z = pc[:,2]
plt.scatter(x,y)
plt.title("X and Y coordinate as Z is zero")
plt.show()
vx = pc[:,6]
vy = pc[:,7]
print(vx,vy)
print(x,y)
ax = plt.axes(projection='3d')

# Data for a three-dimensional line
zline = 0 
xline = vx
yline = vy
ax.scatter(x, y, z, 'marker = m')
ax.scatter(xline,yline,zline, 'marker = m')
plt.title("Velocity and XY coordinate in 3D")
```
